```python
import pandas as pd
examples = []
count = 1
with open("examples/price_big", "r") as file:
    for line in file.readlines():
        examples.append([count, line, 1])
        count+=1

with open("examples/not_price_big", "r") as file:
    for line in file.readlines():
        examples.append([count, line, 0])
        count+=1
```

```python
examples
```

```python
df = pd.DataFrame(examples, columns=["id","example", "positive"])
df = df.set_index("id")
```

```python
df.to_csv("examples/df/price.csv")
```
